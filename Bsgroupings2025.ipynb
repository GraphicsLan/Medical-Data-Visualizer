```python
num_groups = 29
```

```python
# Shuffle the dataset
up = up.sample(frac=1, random_state=42).reset_index(drop=True)

# Assign equal-sized groups
up['Group'] = up.index % num_groups + 1

# Preview the grouped dataset
up

```

```python
# Define the sample size per group
sample_size = 5 # Choose 2 samples per group

# Randomly sample from each group
equal_samples = up.groupby('Group').sample(n=sample_size, random_state=42)

# Display result
equal_samples
```

```python
pd.value_counts(equal_samples['Group'])
```

```python
equal_samples.to_csv('Upschool_grouped.csv', index=False)

```
