```python
df['overweight']=0
```

```python
persons_Bmi=df['weight']/(df['height']/100)**2
```

```python
df['overweight']=persons_Bmi.apply(lambda x:1 if x>25 else 0)
```

```python
df
```

```python
df[['gluc']]=df[['gluc']].applymap(lambda x:0 if x ==1 else 1)
```

```python
df[['cholesterol']]=df[['cholesterol']].applymap(lambda x:0 if x ==1 else 1)
```

```python
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

```python
df_cat = pd.melt(df, id_vars=[], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
```

```python
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])

# Group by 'cardio', 'variable', and 'value', then count occurrences
df_cat = df_cat.groupby(['cardio', 'variable', 'value']).size().reset_index(name='count')
```

```python
A=sns.catplot(data=df_cat, x='variable', y='count', hue='value', col='cardio', kind='bar', palette='Set1')
A.set_axis_labels("Feature", "Count")
A.set_titles("Cardio = {col_name}")
A.set_xticklabels(rotation=30)
A.fig.suptitle("Categorical Feature Distribution by Cardio", fontsize=14, weight='bold')
A.tight_layout();
```

```python
plt.savefig("catplot_by_cardio.png", dpi=300, bbox_inches="tight")  # Save as PNG
```

```python
df_heat = (df['ap_lo'] <= df['ap_hi']),(df['height'] <= df['height'].quantile(0.025)), (df['height'] >= df['height'].quantile(0.975))
```

```python
df_heat=(df['weight']  <= df['weight'].quantile(0.025)), (df['weight'] >= df['weight'].quantile(0.975))
```

```python
df_heat = pd.DataFrame(df)
df_heat = df_heat[
    (df_heat['ap_lo'] <= df_heat['ap_hi']) &  # Valid blood pressure
    (df_heat['height'] >= df_heat['height'].quantile(0.025)) &  # Height lower bound
    (df_heat['height'] <= df_heat['height'].quantile(0.975)) &  # Height upper bound
    (df_heat['weight'] >= df_heat['weight'].quantile(0.025)) &  # Weight lower bound
    (df_heat['weight'] <= df_heat['weight'].quantile(0.975)) ]   # Weight upper bound
df_heat
```

```python
corr=df_heat.corr()
```

```python
mask=np.triu(np.ones_like(corr, dtype=bool))
```

```python
plt.figure(figsize=(8, 6))
```

```python
plt.title("Correlation Matrix Heatmap ", fontsize=16, weight='bold', color='Red')
sns.heatmap(
    corr, mask=mask, annot=True, fmt=".2f", cmap="YlGnBu",  # Different color palette
    vmax=1, vmin=-1, center=0, square=True, linewidths=1.7,  # Thicker borders
    cbar_kws={"shrink": 0.5}, annot_kws={'size': 12, 'weight': 'bold'},  # Adjust annotations
    cbar=True  # Show color bar
)
plt.savefig("heatmap.png", dpi=300, bbox_inches="tight")  # Save as PNG
```
